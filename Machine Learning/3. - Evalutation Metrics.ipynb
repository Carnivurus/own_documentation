## Accuracy

El accuracy es la proporción de predicciones correctas sobre el total de predicciones. Es útil para evaluar la precisión global de un modelo de clasificación, indicando qué tan bien clasifica las instancias positivas y negativas.

$$ \text{Accuracy=} \frac{\text{True Positives + True Negatives}}{\text{TP + TN + FP + FN}}
$$

```python
from sklearn.metrics import accuracy_score

# Supongamos que tienes tus etiquetas verdaderas y predichas
y_true = [1, 0, 1, 0]
y_pred = [1, 0, 1, 1]

# Calcula el accuracy
accuracy = accuracy_score(y_true, y_pred)
print("Accuracy:", accuracy)
```

## Precision

 Indica la proporción de las instancias clasificadas como positivas que `realmente son positivas.`

 $$ \frac{\text{True Positives}}{\text {True Positives + False Positives}} $$

```python
from sklearn.metrics import precision_score

# Supongamos que tienes tus etiquetas verdaderas y predichas
y_true = [1, 0, 1, 0]
y_pred = [1, 0, 1, 1]

# Calcula la precisión
precision = precision_score(y_true, y_pred)
print("Precision:", precision)
```

## Recall

Indica la proporción de las instancias positivas que han sido correctamente `identificadas por el modelo`.

$$ \frac{\text{True Positives}}{\text {True Positives + False Negatives}} $$

```python
from sklearn.metrics import recall_score

# Supongamos que tienes tus etiquetas verdaderas y predichas
y_true = [1, 0, 1, 0]
y_pred = [1, 0, 1, 1]

# Calcula el recall
recall = recall_score(y_true, y_pred)
print("Recall:", recall)
```

## F1 SCore

Is the armonic mean of Precision and Recall.

$$ \frac {2}{\text{(1/Precision)+(1/Recall)}}
$$

```python
from sklearn.metrics import f1_score

# Supongamos que tienes tus etiquetas verdaderas y predichas
y_true = [1, 0, 1, 0]
y_pred = [1, 0, 1, 1]

# Calcula f1_score
f1_score = f1_score(y_true, y_pred)
print("f1_score:", f1_score)
```

## ROC Curve

La curva ROC (Receiver Operating Characteristic) es una representación gráfica de la relación entre la tasa de verdaderos positivos (sensibilidad) y la tasa de falsos positivos (1 - especificidad) para diferentes umbrales de clasificación en un modelo de clasificación binaria.

El área bajo la curva ROC (AUC-ROC) es una medida de la capacidad de discriminación global del modelo: cuanto mayor sea el AUC-ROC, mejor será el rendimiento del modelo en términos de distinguir entre las clases positivas y negativas.

`La curva ROC es una herramienta visual poderosa para evaluar y comparar el rendimiento de modelos de clasificación binaria, especialmente en términos de sensibilidad y especificidad a diferentes umbrales de clasificación.`

```python
import numpy as np
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# Generar datos ficticios para las etiquetas verdaderas y los puntajes de clasificación
np.random.seed(42)
y_true = np.random.randint(0, 2, size=100)  # 100 etiquetas verdaderas (0 o 1)
y_scores = np.random.rand(100)  # Puntajes de clasificación aleatorios entre 0 y 1

# Calcular la curva ROC
fpr, tpr, thresholds = roc_curve(y_true, y_scores)

# Calcular el área bajo la curva ROC (AUC-ROC)
auc = roc_auc_score(y_true, y_scores)

# Graficar la curva ROC
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label='ROC curve (AUC = {:.2f})'.format(auc))
plt.plot([0, 1], [0, 1], color='gray', linestyle='--', label='Random guess')
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend()
plt.grid(True)
plt.show()
```

## MCC (Mathews Correlation Coefficient)

Este modelo estadístico evalua todos los positivos y negativos obtenidos (Falsos y Verdaderos). Hesta le da el mismo peso a los Verdaderos Negativos y los Verdaderos positivos.

Nota: Solo funciona para predicciones binarias.

```python
from sklearn.metrics import matthews_corrcoef
y_true = [1, 0, 1, 0]
y_pred = [1, 0, 1, 1]
mcc = matthews_corrcoef(y_true, y_pred)
print("Coeficiente de correlación Matthews (MCC):", mcc)
```

## Root Mean Squared Error

El "Mean Squared Error" (MSE) es una medida comúnmente utilizada para evaluar la precisión de un modelo de regresión. Mide el promedio de los cuadrados de las diferencias entre los valores predichos por el modelo y los valores reales. Sin embargo, el "Root Mean Squared Error" (RMSE) es una variante del MSE que toma la raíz cuadrada del valor resultante, lo que proporciona una medida de error en las mismas unidades que la variable de destino original, haciéndola más interpretable.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np

# Supongamos que tienes tus datos de entrenamiento X_train y y_train
X_train = np.array([[1], [2], [3], [4], [5]])
y_train = np.array([2, 4, 6, 8, 10])

# Entrenar el modelo de regresión lineal
model = LinearRegression()
model.fit(X_train, y_train)

# Supongamos que tienes tus datos de prueba X_test y y_test
X_test = np.array([[6], [7], [8], [9], [10]])
y_test = np.array([12, 14, 16, 18, 20])

# Realizar predicciones en los datos de prueba
y_pred = model.predict(X_test)

# Calcular el RMSE
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print("Root Mean Squared Error (RMSE):", rmse)
```
