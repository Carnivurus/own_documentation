## Extraer datos a través de HTML

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd 

'''The following code will connect to a webpage to extract data from a table, the result will be a DataFrame'''

response =requests.get('https://practicum-content.s3.us-west-1.amazonaws.com/data-analyst-eng/moved_chicago_weather_2017.html')
soup = BeautifulSoup(response.text,'lxml')

header = []
content = []

table = soup.find('table',{'id':"weather_records"}) #Need to be treated as a single element not find_all
for row in table.find_all('th'): #Now we can use all the find all to look for each elemnt
    header.append(row.text)

for row in table.find_all('tr'):
   if not row.find_all('th'):
      content.append([element.text for element in row.find_all('td')])

weather_records = pd.DataFrame(data=content, columns=header)

print(weather_records)
```
