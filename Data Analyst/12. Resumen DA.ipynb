# Medidas Stadisticas de dispersion

```python
import numpy as np
import pandas as pd
import math
import random as rand
import matplotlib.pyplot as plt
```

- Varianza $\sigma^2 $  = La varianza es: la suma promedio de los cuadrados de las desviaciones y describe que tan dispersos estan los valores con respecto a su media

- Desviación standard $\sigma $  = La desviación estándar es simplemente la raíz cuadrada de la varianza. Proporciona una medida de dispersión que tiene las mismas unidades que los datos originales, lo que facilita su interpretación

- Covarianza = La covarianza es una medida de la relación lineal entre `dos variables.` Si la covarianza es positiva, indica que las variables tienden a aumentar o disminuir juntas; si es negativa, indica que cuando una variable aumenta, la otra tiende a disminuir. Si la covarianza es cero, no hay una relación lineal

```python
# Definimos dos listas para obtener varianza, covarianza y desviación standard.
array= [rand.randint(1,15) for _ in range(50)]
array2= [rand.randint(1,15) for _ in range(50)]
 
mean = np.mean(array)
mean2 = np.mean(array2)
varianza = np.var(array, ddof=1) #ddof es para delimitar si lo aplicamos a una muestra o una población.

# La desviación standard es la raiz cuadrada de la varianza
desviación_standard = math.sqrt(varianza) # Aunque el resultado es cercano, es preferible usar el metodo np.std()
desviación_standard_directa = np.std(array) # Es la medida de separación de una lista de datos ordenada

# La covarianza es simplemente la relación de varianzas entre dos o mas variables
covarianza = np.cov(array,array2) 

print(f'varianza: {varianza:.2f}')
print(f'desviación standard: {desviación_standard:.2f}')
print(f'desviación standard directa: {desviación_standard_directa:.2f}')
print(f'Covarianza: {covarianza}')
```

Con esto tenemos una $\sigma^2$ de 9.17, que para podes visualizarla mejor la pasamos a $\sigma$ con 3.03.

```python
#Creando histograma
plt.hist(array, bins=10, color='lightblue', label='Frecuencia')

#Añadiendo lineas al grafico
plt.axvline(mean, color='r', label='Media')
plt.axvline(mean + desviación_standard, color='green', label='Desviacion_derecha')
plt.axvline(mean - desviación_standard, color='green', label='Desviacion_izquierda')

#Agregando etiquetas
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Histograma con media y desviación')
plt.legend()


plt.show()
```

```python
plt.scatter(array,array2, color= 'lightblue')
plt.title('Diagrama de Dispersión con Covarianza')
plt.xlabel('Variable 1')
plt.ylabel('Variable 2')

# Mostramos la covarianza en la gráfica.
# plt.text(4, 3, f'Covarianza = {covarianza}', fontsize=12, color='red')


plt.show()
```

De esta manera podemos ver gráficamente como la desviación representa la distribución conforme a la media.

# Permutaciones

Las permutaciones se refieren al número de formas diferentes en que se pueden ordenar los elementos de un conjunto, Una permutación de un conjunto de elementos es una disposición ordenada de esos elementos. Por ejemplo, si tienes un conjunto de 𝑛 elementos, el número total de permutaciones de esos elementos es 𝑛!

`Una permutación es un arreglo donde el orden no es importante`

Permutaciones de ABC: (ABC, ACB, BAC, BCA, CAB, CBA)

```python
import math
from math import factorial as f

n=3 # Numero de letras
k=3 # Conjunto de letras que necesito (agrupado de 3 en 3)
permutaciones = math.perm(n,k)
print(f'Se tienen {permutaciones} combinaciones')
```

# Distribución binomial

La distribución binomial es un modelo estadístico que describe `el número de éxitos en una secuencia de ensayos independientes`,  donde cada ensayo tiene dos posibles resultados: éxito o fracaso 

Yo resumo la formula en: el numero de combinaciones posibles por la probabilidad de que ocurran los eventos exitosos por la probabilidad de los eventos que no ocurrirán

```python
from scipy.stats import binom

n = 100 #veces que se tiró una moneda
k = 50 # veces que esperamos tener exito
p = 0.5 #probabilidad de que ocurra cada evento único
probability = binom.pmf(k,n,p)


# El resultado será el resultado de la multiplicación de todas las probabilidades individuales
print(f'La probabilidad de es de {round(probability*100,2)}%')
```

# Distribución Normal

La distribución normal, también conocida como distribución gaussiana, es una de las distribuciones de probabilidad más importantes en estadística y teoría de probabilidad. La Moda, Mediana y Media son la misma al tener una distribución normal cuando esta estandarizada


Su formula es la siguiente: $$X \sim (μ, σ^2)$$

- X = variable aleatoria
- ~ = 'Sigue una distribución'
- μ = La media
- σ<sup>2</sup> = La desviación estandar

Distribución Normal con Cumulative Distribution Function (Porcentajes)

```python
'''Supongamos que el número de horas que los clientes pasan en un gimnasio cada semana sigue
 una distribución normal con una media de 2 y una desviación típica de 1. 
 ¿Qué porcentaje de clientes pasa más de 4 horas a la semana en el gimnasio?
 ¿Qué porcentaje de clientes pasa menos de 4 horas a la semana en el gimnasio?
 ¿Qué porcentaje de clientes pasa exactamente de 4 horas a la semana en el gimnasio?
 '''

from scipy import stats

mu = 2  # Media (horas)
sigma = 1  # Desviación estándar

# Definir el límite de horas para cada caso
horas_limite = 4

# Calcular la probabilidad acumulada hasta el valor del límite de horas
prob_limite_horas = stats.norm.cdf(horas_limite, mu, sigma)

# La probabilidad de que los clientes pasen más de 4 horas es el complemento de la probabilidad de que pasen hasta 4 horas
prob_mas_de_4_horas = 1 - prob_limite_horas

# La probabilidad de que los clientes pasen menos de 4 horas es simplemente la probabilidad acumulada hasta 4 horas
prob_menos_de_4_horas = prob_limite_horas

# La probabilidad de que los clientes pasen exactamente 4 horas es la diferencia entre las probabilidades acumuladas hasta 4 horas y hasta un poco menos de 4 horas
prob_exactamente_4_horas = prob_limite_horas - stats.norm.cdf(horas_limite - 0.001, mu, sigma)

# Convertir las probabilidades en porcentajes
porcentaje_mas_de_4_horas = prob_mas_de_4_horas * 100
porcentaje_menos_de_4_horas = prob_menos_de_4_horas * 100
porcentaje_exactamente_4_horas = prob_exactamente_4_horas * 100

# Imprimir los resultados
print(f'Porcentaje de clientes que pasa más de 4 horas: {porcentaje_mas_de_4_horas:.2f}%')
print(f'Porcentaje de clientes que pasa menos de 4 horas: {porcentaje_menos_de_4_horas:.2f}%')
print(f'Porcentaje de clientes que pasa exactamente 4 horas: {porcentaje_exactamente_4_horas:.2f}%')
```
