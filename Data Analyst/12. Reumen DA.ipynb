# Medidas Stadisticas de dispersion

```python
import numpy as np
import pandas as pd
import math
import random as rand
import matplotlib.pyplot as plt
```

- Varianza $\sigma^2 $  = La varianza es: la suma promedio de los cuadrados de las desviaciones y describe que tan dispersos estan los valores con respecto a su media

- Desviación standard $\sigma $  = La desviación estándar es simplemente la raíz cuadrada de la varianza. Proporciona una medida de dispersión que tiene las mismas unidades que los datos originales, lo que facilita su interpretación

- Covarianza = La covarianza es una medida de la relación lineal entre `dos variables.` Si la covarianza es positiva, indica que las variables tienden a aumentar o disminuir juntas; si es negativa, indica que cuando una variable aumenta, la otra tiende a disminuir. Si la covarianza es cero, no hay una relación lineal

```python
# Definimos dos listas para obtener varianza, covarianza y desviación standard.
array= [rand.randint(1,15) for _ in range(50)]
array2= [rand.randint(1,15) for _ in range(50)]
 
mean = np.mean(array)
mean2 = np.mean(array2)
varianza = np.var(array, ddof=1) #ddof es para delimitar si lo aplicamos a una muestra o una población.

# La desviación standard es la raiz cuadrada de la varianza
desviación_standard = math.sqrt(varianza) # Aunque el resultado es cercano, es preferible usar el metodo np.std()
desviación_standard_directa = np.std(array) # Es la medida de separación de una lista de datos ordenada

# La covarianza es simplemente la relación de varianzas entre dos o mas variables
covarianza = np.cov(array,array2) 

print(f'varianza: {varianza:.2f}')
print(f'desviación standard: {desviación_standard:.2f}')
print(f'desviación standard directa: {desviación_standard_directa:.2f}')
print(f'Covarianza: {covarianza}')
```

Con esto tenemos una $\sigma^2$ de 9.17, que para podes visualizarla mejor la pasamos a $\sigma$ con 3.03.

```python
#Creando histograma
plt.hist(array, bins=10, color='lightblue', label='Frecuencia')

#Añadiendo lineas al grafico
plt.axvline(mean, color='r', label='Media')
plt.axvline(mean + desviación_standard, color='green', label='Desviacion_derecha')
plt.axvline(mean - desviación_standard, color='green', label='Desviacion_izquierda')

#Agregando etiquetas
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Histograma con media y desviación')
plt.legend()


plt.show()
```

```python
plt.scatter(array,array2, color= 'lightblue')
plt.title('Diagrama de Dispersión con Covarianza')
plt.xlabel('Variable 1')
plt.ylabel('Variable 2')

# Mostramos la covarianza en la gráfica.
# plt.text(4, 3, f'Covarianza = {covarianza}', fontsize=12, color='red')


plt.show()
```

De esta manera podemos ver graficamente como la desvisación representa la distribución conforme a la media.
